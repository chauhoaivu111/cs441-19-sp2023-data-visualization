```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# the URL of the website
url = 'https://www.worldometers.info/coronavirus/'


response = requests.get(url)


if response.status_code == 200:
 
    soup = BeautifulSoup(response.content, 'html.parser')

    
    table = soup.find('table', {'id': 'main_table_countries_today'})

    headers = []
    for th in table.find_all('th'):
        headers.append(th.text.strip())

    data = []
    for tr in table.tbody.find_all('tr'):
        row = []
        for td in tr.find_all('td'):
            row.append(td.text.strip())
        data.append(row)

    df = pd.DataFrame(data, columns=headers)

    # replace "N/A" values with NaN
    df.replace("N/A", pd.np.nan, inplace=True)

    
    numeric_columns = ["TotalCases", "NewCases", "TotalDeaths", "NewDeaths", "TotalRecovered", "NewRecovered", "ActiveCases", "TotalTests"]
    df[numeric_columns] = df[numeric_columns].apply(lambda x: pd.to_numeric(x.str.replace(',', '')))

    # get the DataFrame to a CSV file
    df.to_csv('covid19_stats.csv', index=False)
else:
    print('Request failed with status code', response.status_code)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('covid19_stats.csv')


continents = ["North America", "Asia", "Europe", "South America", "Oceania"]
grouped_data = df[df["Country,Other"].isin(continents)].groupby("Country,Other")["TotalCases"].sum().sort_values(ascending=True)

plt.figure(figsize=(8, 5))
plt.barh(grouped_data.index, grouped_data.values, color='royalblue')
plt.title('Total COVID-19 Cases by Continent')
plt.xlabel('Total Cases')
plt.ylabel('Continent')
plt.xticks([])
plt.show()
```

# New Deaths from COVID-19 in Asia

```python

# filter the data to only include rows where the 'Continent' column is 'Asia'
asia_data = df[df['Continent'] == 'Asia']
plt.figure(figsize=(10, 6))
plt.bar(asia_data['Country,Other'], asia_data['NewDeaths'])
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('New Deaths')
plt.title('New Deaths from COVID-19 in Asia')
plt.show()
```

# New Cases from COVID-19 in Asia

```python
# filter the data to only include rows where the 'Continent' column is 'Asia'
asia_data = df[df['Continent'] == 'Asia']
asia_data['NewCases'] = pd.to_numeric(asia_data['NewCases'], errors='coerce')
plt.figure(figsize=(10, 6))
plt.bar(asia_data['Country,Other'], asia_data['NewCases'])
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('New Cases')
plt.title('New Cases from COVID-19 in Asia')
plt.show()
```

# Top 10 Countries with the Most COVID-19 Cases in Asia

```python

asia_data = df[df['Continent'] == 'Asia']

# Sort the data by the 'TotalCases' column in descending order and select the top 10 rows
top_10_countries = asia_data.sort_values(by='TotalDeaths', ascending=False).head(10)

plt.bar(top_10_countries['Country,Other'], top_10_countries['TotalDeaths'])
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Total Deaths')
plt.title('Top 10 Countries with the Most COVID-19 Cases in Asia')
plt.show()
```

```python
import pandas as pd
import folium
from random import sample

df = pd.read_csv('dublin.csv')

random_indices = sample(range(len(df)), 10)

# Create a new dataframe with only the randomly selected rows
random_df = df.iloc[random_indices]

# the map at the center of Dublin
map = folium.Map(location=[53.349562,-6.278198], zoom_start=12)

# Add a marker for each random point to the map
for index, row in random_df.iterrows():
    # Check if latitude and longitude values are missing or invalid
    if pd.notnull(row['Latitude']) and pd.notnull(row['Longitude']):
        folium.Marker([row['Latitude'], row['Longitude']]).add_to(map)

# Display the map
map
```
